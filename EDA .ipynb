# Encoding

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
Data['sex'] = LE.fit_transform(Data['sex'])
Data['steroid'] = LE.fit_transform(Data['steroid'])
Data['antivirals'] = LE.fit_transform(Data['antivirals'])
Data['fatigue'] = LE.fit_transform(Data['fatigue'])
Data['malaise'] = LE.fit_transform(Data['malaise'])
Data['anorexia'] = LE.fit_transform(Data['anorexia'])
Data['liver_big'] = LE.fit_transform(Data['liver_big'])
Data['spleen_palpable'] = LE.fit_transform(Data['spleen_palpable'])
Data['spiders'] = LE.fit_transform(Data['spiders'])
Data['varices'] = LE.fit_transform(Data['varices'])
Data['histology'] = LE.fit_transform(Data['histology'])
Data['class'] = LE.fit_transform(Data['class'])
```

```python
Data.head()
```

# Corelation

```python
data_corr=Data.corr()
data_corr
```

# Heatmap

```python
fiq=plt.figure(figsize=(20,20))
sb.heatmap(data_corr,annot=True,cmap='RdYlBu')
```

# Visualization

### Barplot

```python
chart=plt.figure(figsize=(20,20))
sb.barplot(x=Data['age'],y=Data['spiders'],hue=Data['class'])
```

### Histogram

```python

f, axes = plt.subplots(2, 2, figsize=(10,10), sharex=True)
sb.distplot( Data["bilirubin"] , color="skyblue", ax=axes[0, 0])
sb.distplot( Data["alk_phosphate"] , color="olive", ax=axes[0, 1])
sb.distplot( Data["sgot"] , color="gold", ax=axes[1, 0])
sb.distplot( Data["albumin"] , color="teal", ax=axes[1, 1])
```

### Swarm plot

```python
chart=plt.figure(figsize=(80,40))
sb.catplot(data=Data, kind="swarm", x="sex", y="sgot", hue="class")
```

```python
sb.relplot(
    data=Data,
    x="age", y="sgot", hue="class"
)
```

```python
sb.countplot(x='class',hue='sex',data=Data)
```
